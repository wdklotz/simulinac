```python
#%matplotlib ipympl
#%matplotlib tk
%matplotlib widget
# %matplotlib notebook

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[9.25,5]
import numpy as np
import math
import h5py
from setutil import Ktp as K

def frame_plot(ax,rdata,cdata,title=None,label=None,xlabel=None,ylabel=None):
    obj, = ax.plot(rdata,cdata,'.',ms=0.8)
    if title  != None: ax.set_title(title)
    if label  != None: obj.set_label(label)
    if xlabel != None: ax.set_xlabel(xlabel)
    if ylabel != None: ax.set_ylabel(ylabel)
    ax.legend(loc='upper right')
    return ax

h5File=h5py.File('30k.h5','r')
frames=datasets=h5File['/frames']
ds_keys=list(datasets.keys())
ds_keys.sort(key=int)
last_frame=ds_keys[-1]
# print(f'last frame {last_frame}')
# print(f'1st record in frame #{last_frame} \n{frames[last_frame][0]}')

# first & last frame
InOutFrames = (frames[ds_keys[0]],frames[last_frame])

# figure & axis
fig, axs = plt.subplots(3,2,constrained_layout=True,figsize=(13,11))
x,xp,y,yp,z,zp = [0,1,2,3,4,5]

# read data sets from h5 file, make scatter plots
for rcnt in range(3):
    for ccnt in range(2):
        pos = (rcnt,ccnt)
        if pos == (0,0):
            frame = InOutFrames[0]
            x_axe=[particle[x] for particle in frame]
            y_axe=[particle[xp] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="x,x'",label="IN",xlabel='m')
            frame = InOutFrames[1]
            x_axe=[particle[x] for particle in frame]
            y_axe=[particle[xp] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="x,x'",label="OUT",xlabel='m')
            continue
        if pos == (0,1):
            frame = InOutFrames[0]
            x_axe=[particle[y] for particle in frame]
            y_axe=[particle[yp] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="y,y'",label="IN",xlabel='m')
            frame = InOutFrames[1]
            x_axe=[particle[y] for particle in frame]
            y_axe=[particle[yp] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="y,y'",label="OUT",xlabel='m')
        if pos == (1,0):
            frame = InOutFrames[0]
            x_axe=[particle[x] for particle in frame]
            y_axe=[particle[y] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="x,y",label="IN",xlabel='m')
            frame = InOutFrames[1]
            x_axe=[particle[x] for particle in frame]
            y_axe=[particle[y] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="x,y",label="OUT",xlabel='m')
        if pos == (1,1):
            frame = InOutFrames[0]
            x_axe=[particle[xp] for particle in frame]
            y_axe=[particle[yp] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="x',y'",label="IN",xlabel='m')
            frame = InOutFrames[1]
            x_axe=[particle[xp] for particle in frame]
            y_axe=[particle[yp] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="x',y'",label="OUT",xlabel='m')
        if pos == (2,0):
            frame = InOutFrames[0]
            x_axe=[particle[z] for particle in frame]
            y_axe=[particle[zp] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="z,z'",label="IN",xlabel='m')
            frame = InOutFrames[1]
            x_axe=[particle[z] for particle in frame]
            y_axe=[particle[zp] for particle in frame]
            frame_plot(axs[rcnt][ccnt],x_axe,y_axe,title="z,z'",label="OUT",xlabel='m')
```
